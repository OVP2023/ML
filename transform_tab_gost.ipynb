```python
#Транспонирование нужных столбцов матрицы с сайта
import pandas as pd
```

```python
df=pd.read_excel('/content/drive/MyDrive/Stajerovka_осень_2024/Парсинг/table_gost_0124.xlsx')
df.rename(columns={0:'Название',1:'Значение'}, inplace=True)
df.info()
```

```python
dff=df[(df['Название']=='Обозначение') | (df['Название']=='Заглавие на русском языке') | (df['Название']=='Дата введения в действие') | (df['Название']=='Дата принятия в МГС') | (df['Название']=='Дата последнего издания') | (df['Название']=='Дата приказа о закреплении документа за ТК')]
print(dff.shape[0])
dff=dff.drop(columns=['Unnamed: 0'])
dff.fillna('-',inplace=True)
dff.info()
```

```python

list_name=dff['Название'].tolist()
list_val=dff['Значение'].tolist()
l1=[]
l2=[]
l3=[]
l4=[]
l5=[]
l6=[]

for ind in range(0,len(list_name)):
    if list_name[ind]=='Обозначение':
        l1.append(list_val[ind])
    elif list_name[ind]=='Заглавие на русском языке':
        l2.append(list_val[ind])
    elif list_name[ind]=='Дата введения в действие':
        l3.append(list_val[ind])
    elif list_name[ind]=='Дата принятия в МГС':
        l4.append(list_val[ind])
    elif list_name[ind]=='Дата последнего издания':
        l5.append(list_val[ind])
    elif list_name[ind]=='Дата приказа о закреплении документа за ТК':
        l6.append(list_val[ind])

print(len(l1),len(l2),len(l3),len(l4),len(l5),len(l6))
print(l6)

df_result=pd.DataFrame({'Обозначение':l1,'Заглавие на русском языке':l2,'Дата введения в действие':l3,'Дата принятия в МГС':l4, 'Дата последнего издания':l5, 'Дата приказа о закреплении документа за ТК':l6})
df_result.head(40)
```

```python
df_result.to_excel('/content/drive/MyDrive/Stajerovka_осень_2024/Парсинг/table_gost_0124V2.xlsx')
```
