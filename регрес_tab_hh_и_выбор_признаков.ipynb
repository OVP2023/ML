```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest,chi2,RFE
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics


df=pd.read_excel('/content/drive/MyDrive/Stajerovka_осень_2024/статистика hh/tab_hh.xlsx')
df.info()
```

```python
X=df
y=X['Средняя зарплата']

X = X.drop(columns=['№№','Название специализации','Средняя зарплата'])
#X = X.dropna()

X.info()
```

```python
from sklearn.pipeline import make_pipeline  # используем пайплайны
from sklearn.preprocessing import StandardScaler

model = RandomForestRegressor(max_depth=5,min_samples_split=2, random_state=10)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
model.fit(X_train, y_train)
print(round(model.score(X_test,y_test),3))
```

```python
model3 = make_pipeline(
    #StandardScaler(),
    MinMaxScaler(),
    RandomForestRegressor(max_depth=5,min_samples_split=4, random_state=10)
)

X_train, X_test, y_train, y_test = train_test_split(XXX, y, test_size=0.3, random_state=0)
model3.fit(X_train, y_train)
print(round(model3.score(X_test,y_test),3))
```
